```python
class Person:
    
    def __init__(self, name, residence):
        self.name = name
        self.residence = residence
        
    def __repr__(self):
        return f'Person("{self.name}", "{self.residence}")'
        
    def tell(self):
        return 'I am %s and I live in %s' % (self.name, self.residence)

    def move(self, new_residence):
        self.residence = new_residence
        
```

```python
p = Person('Peter', 'Utrecht')
```

```python
print(p.tell())
```

```python
p.move('Amsterdam')
print(p.tell())
```

```python
groep = []

groep.append(Person('Peter', 'Utrecht'))
groep.append(Person('Janneke', 'Amsterdam'))
groep.append(Person('Kim', 'Delft'))

groep
```

```python
namen = ['Peter', 'Janneke', 'Kim']

for naam in namen:
    print(naam)
```

```python
for persoon in groep:
    print(persoon.tell())
```
