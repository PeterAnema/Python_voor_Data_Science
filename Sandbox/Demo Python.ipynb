## Functies

```python
def groet():
    print('Hoi, Hoe is het?')
    print('Ik ben echt benieuwd.')
    print('Met mij gaat het goed.')
```

```python
groet()
```

```python
def groet(naam):
    print(f'Hoi {naam}, Hoe is het?')
    print('Ik ben echt benieuwd.')
    print('Met mij gaat het goed.')
```

```python
groet('Peter')
```

```python
def groet(naam, herhaal):
    print(f'Hoi {naam}, Hoe is het?')
    for i in range(herhaal):
        print('Ik ben echt benieuwd.')
    print('Met mij gaat het goed.')
```

```python
groet('Peter', 1)
```

```python
def groet(naam, herhaal = 1):
    print(f'Hoi {naam}, Hoe is het?')
    for i in range(herhaal):
        print('Ik ben echt benieuwd.')
    print('Met mij gaat het goed.')
```

```python
groet('Peter')
```

```python
voornaam = 'Peter'
groet(voornaam)
```

```python
groet(herhaal = 10, naam = 'Peter')
```

```python
def telop_en_vermenigvuldig(getal1, getal2, factor = 1):
    resultaat = (getal1 + getal2) * factor
    return resultaat
```

```python
telop_en_vermenigvuldig(4, 7)
```

```python
telop_en_vermenigvuldig(4, 7, factor = 3)
```
