## Aggregatie

```python
df_mean_weights = df[['origin','weight']].groupby('origin').mean()
df_mean_weights 
```

```python
df.groupby('origin')[['weight','horsepower']].mean()
```

```python
totalen = df.groupby('origin')['weight'].agg(['min','max','mean','median'])
totalen
```

```python
df.merge(totalen, left_on='origin', right_index=True)
```

```python
df.groupby('origin', as_index=False)['cylinders'].agg(lambda a: ','.join(map(str, sorted(set(a)))))
```

```python
pd.pivot_table(df,
               index='origin', 
               columns='cylinders',
               values='mpg', 
               aggfunc=['min','max','mean']).swaplevel(0, 1, axis = 1).sort_index(axis = 1)
```

```python
df['make'] = df['model'].str.extract(r'^(\w+)\s.*')
```

```python
df.groupby('make', as_index=False)['cylinders'].agg(lambda a: ','.join(map(str, sorted(set(a)))))
```
